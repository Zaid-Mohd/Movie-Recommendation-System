    1- Content Based :- On the Basis of similarity of content 
    2- Collebrative Filtering :- On the Basis of user similarity

```python
# Importing ast library for coverting string list into list
import ast
```

```python
# Making function for fetching name that we need from dictionaries
def convert(text):
    L = []
    text = ast.literal_eval(text)
    for i in text :
        L.append(i['name']) 
    return L
```

```python
# appling funtion convert 
movie['genres'] = movie['genres'].apply(convert)
```

```python
# checking genres column after applying function
movie['genres']
```

```python
# appling function convert on keywords
movie['keywords'] = movie['keywords'].apply(convert)
movie['keywords']
```

```python
# detail of first row of cast column 
movie['cast'][0]
```

```python
# function for fecting top three cast from the dictionaries
def convert_cast(text):
    L = []
    count = 0
    text = ast.literal_eval(text)
    for i in text :
        if count != 3 :
            L.append(i['name']) 
            count = count + 1
        else :
            break
    return L
```

```python
# appling funtion to fetched the top 3 cast
movie['cast'] = movie['cast'].apply(convert_cast)
```

```python
# appling funtion to fetched
movie['cast']
```

```python
ast.literal_eval(movie['crew'][0])[0]
```

```python
# function for fecting director and music composer name from crew column 
def fetch_director(text):
    L = []
    text = ast.literal_eval(text)
    for i in text :
        if i['job'] == 'Director' :
            L.append(i['name'])
        if i['job'] == 'Original Music Composer' :
            L.append(i['name']) 
    return L
```

```python
# appling funtion to fetched
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
# Checking data after applying functions
movie
```

```python
# Now Converting overview (String Data) column into List
L = []
for i in movie['overview'] :
    L.append(i.split())
```

```python
movie['overview'] = L
```

```python
movie
```

```python
# removing space from the words 
movie['cast'].apply(lambda x: [i.replace(" " , "") for i in x])
```

```python
# removing space for treating words properly (For treating single word)
def remove_space(data) :
    L = []
    for i in data :
        L.append(i.replace(" " , ""))
    return L
```

```python
# appling function
movie['genres'] = movie['genres'].apply(remove_space)
movie['keywords'] = movie['keywords'].apply(remove_space)
movie['cast'] = movie['cast'].apply(remove_space)
movie['crew'] = movie['crew'].apply(remove_space)
movie
```

```python
# making final data to work and making a column named tags by adding genres , keywords , cast , crew and overview
# Then will find recommendation on the bases of tags
Final_Data = movie[['movie_id' , 'title']] 

Final_Data['tags'] = movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew'] + movie['overview']
```

```python
# Final data to work on 
Final_Data
```

```python
Final_Data['tags'] = Final_Data['tags'].apply(lambda x : " ".join(x))
Final_Data['tags']
```

```python
# importing nltk for appling Stammering on tags columns 
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
# function for Stammering
def stem(data) :
    L= []
    for i in data.split() :
        L.append(PorterStemmer().stem(i))
    return " ".join(L)
```

```python
# appling function on tags to remove similar works
Final_Data['tags'] = Final_Data['tags'].apply(stem)
Final_Data['tags']
```

```python
# importing CountVectorizer function to convert tags into text vectors
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# words count is 2000 and also removing stop words 
CV = CountVectorizer(max_features = 2000 , stop_words = 'english')
```

```python
# finally coverted tags columns into vectors
vectors = CV.fit_transform(Final_Data['tags'])
vectors = vectors.toarray()
vectors
```

```python
vectors[0]
```

```python
# list of 2000 words that are used to make vectors
CV.get_feature_names_out().tolist()
```

```python
# importing cosine similarity function to calculate the distace of vector
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# finding similarity
# 1 means movie is exectly same and 0 means movies are opposite
similarity = cosine_similarity(vectors)
similarity
```

```python
Final_Data[Final_Data['title'] == 'Avatar'].index[0] # Fining Index
```

```python
# sorting similarity for fetching top 5 similar movies
# using enumerate for keeping index posotions so that we can fetched the movies name with index
sorted(list(enumerate(similarity[0])) , reverse = True , key = lambda X: X[1])[1:6]
```

```python
# Making function and printing name of top 5 similar movies
def recommend(movie) :
    movie_index = Final_Data[Final_Data['title'] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[movie_index])) , reverse = True , key = lambda X: X[1])[1:6]
    
    for i in movies_list :
        print(Final_Data.iloc[i[0]].title)
```

```python
recommend('The Avengers')
```

```python
recommend('Aliens vs Predator: Requiem')
```

```python
recommend('Edge of Tomorrow')
```
